```python
# ============================================
# HYPOTHESIS TESTING — SIMPLE PPT VERSION
# ============================================

import pandas as pd
import numpy as np
from scipy import stats

# Load Dataset
df = pd.read_excel("Food_delivery_dataset.xlsx")

# =====================================================
# TEST 1 — ONE SAMPLE Z-TEST
# Mean Delivery Time vs 36 mins
# =====================================================

delivery = df['Delivery_Time_Minutes'].dropna()

mean_delivery = delivery.mean()
std_delivery = delivery.std()
n = len(delivery)

# Z-Test
z_stat = (mean_delivery - 36) / (std_delivery / np.sqrt(n))
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("TEST 1 — Z TEST")
print("Z-Statistic :", round(z_stat,2))
print("P-Value :", round(p_value,5))

if p_value < 0.05:
    print("Decision : REJECT H0")
else:
    print("Decision : FAIL TO REJECT H0")

print("Mean Delivery :", round(mean_delivery,2))


# =====================================================
# TEST 2 — TWO SAMPLE T-TEST
# Festival vs Non-Festival Order Value
# =====================================================

festival = df[df['Festival_Flag']=='FESTIVAL']['Order_Value']
non_festival = df[df['Festival_Flag']=='NON-FESTIVAL']['Order_Value']

t_stat, p_value = stats.ttest_ind(festival, non_festival)

print("\nTEST 2 — T TEST")
print("T-Statistic :", round(t_stat,2))
print("P-Value :", round(p_value,5))

if p_value < 0.05:
    print("Decision : REJECT H0")
else:
    print("Decision : FAIL TO REJECT H0")

print("Festival Avg :", round(festival.mean(),2))



# =====================================================
# TEST 2B — PAIRED T-TEST
# Order Value vs Final Amount
# =====================================================

# H0 :order value have no impact on final amount
# H1 : order value  reduce the final amount

paired = df[['Order_Value', 'Final_Amount']].dropna()

t_stat, p_value = stats.ttest_rel(
    paired['Order_Value'],
    paired['Final_Amount']
)

print("TEST 2B — PAIRED T TEST")
print("T-Statistic :", round(t_stat,2))
print("P-Value :", round(p_value,5))

if p_value < 0.05:
    print("Decision : REJECT H0")
else:
    print("Decision : FAIL TO REJECT H0")


# =====================================================
# TEST 3 — CHI-SQUARE TEST OF INDEPENDENCE
# Payment Mode vs Customer Type
# =====================================================

table = pd.crosstab(df['Customer_Type'], df['Payment_Mode'])

chi2, p_value, dof, expected = stats.chi2_contingency(table)

print("\nTEST 3 — CHI SQUARE")
print("Chi2 Value :", round(chi2,2))
print("P-Value :", round(p_value,5))

if p_value < 0.05:
    print("Decision : REJECT H0")
else:
    print("Decision : FAIL TO REJECT H0")
```

```python
# =====================================================
# TEST 4A — CHI-SQUARE GOODNESS OF FIT
# =====================================================

# H0 : Ratings are equally distributed
# H1 : Ratings are NOT equally distributed

ratings = df['Order_Rating'].value_counts().sort_index()

expected = np.full(len(ratings), len(df)/len(ratings))

chi2_gof, p_value = stats.chisquare(ratings, expected)

print("TEST 4A — CHI SQUARE GOF")
print("Chi2 Value :", round(chi2_gof,2))
print("P-Value :", round(p_value,5))

if p_value < 0.05:
    print("Decision : REJECT H0")
else:
    print("Decision : FAIL TO REJECT H0")


# =====================================================
# TEST 4B — CHI-SQUARE HOMOGENEITY TEST
# =====================================================

# H0 : Rating distribution is same across regions
# H1 : Rating distribution differs across regions

table = pd.crosstab(df['Region'], df['Order_Rating'])

chi2, p_value, dof, expected = stats.chi2_contingency(table)

print("\nTEST 4B — HOMOGENEITY TEST")
print("Chi2 Value :", round(chi2,2))
print("P-Value :", round(p_value,5))

if p_value < 0.05:
    print("Decision : REJECT H0")
else:
    print("Decision : FAIL TO REJECT H0")
```

* **Test 1 (Z-Test):** Delivery time is significantly different from the 36-minute benchmark.

* **Test 2A (Independent T-Test):** Festival periods have significantly higher order values than non-festival periods.

* **Test 2B (Paired T-Test):** Discounts significantly reduce the final amount paid by customers.

* **Test 3 (Chi-Square Independence):** Payment mode and customer type are independent of each other.

* **Test 4A (Chi-Square Goodness of Fit):** Customer ratings are not equally distributed across all rating levels.

* **Test 4B (Chi-Square Homogeneity):** Customer rating distribution is similar across all regions.
