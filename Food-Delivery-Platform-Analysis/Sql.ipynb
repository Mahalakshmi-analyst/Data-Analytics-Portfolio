```python
import pandas as pd
from sqlalchemy import create_engine

# Load Excel file
df = pd.read_excel("Food_delivery_dataset.xlsx")

# MySQL credentials
user = "root"
password = "12345678"  
host = "localhost"
port = 3306
database = "food_delivery"

# Create connection
engine = create_engine(
    f"mysql+pymysql://{user}:{password}@{host}:{port}/{database}"
)

# Upload to MySQL
df.to_sql(
    name="main_table",
    con=engine,
    if_exists="replace",
    index=False,
    chunksize=10000   # important for large files
)

print("Data Import Done")
```
