```python
# Load 
import pandas as pd
import numpy as np
```

```python
# Load Data
customers = pd.read_excel("Customers.xlsx")
delivery = pd.read_excel("Delivery_Partners.xlsx")
orders = pd.read_excel("Orders.xlsx")
restaurants = pd.read_excel("Restaurants.xlsx")
```

```python
# CHECK DATA SHAPE
print("Customers:", customers.shape)
print("Delivery:", delivery.shape)
print("Orders:", orders.shape)
print("Restaurants:", restaurants.shape)
```

```python
#  CHECK NULL VALUES
print("Customers:  \n", customers.isnull().sum())
print("Delivery:  \n", delivery.isnull().sum())
print("Orders:    \n", orders.isnull().sum())
print("Restaurants:   \n", restaurants.isnull().sum())
```

```python
print("Customers:  \n", customers.dtypes)
print("Delivery:  \n", delivery.dtypes)
print("Orders:    \n", orders.dtypes)
print("Restaurants:   \n", restaurants.dtypes)
```

```python
orders["Order_Rating"] = orders["Order_Rating"].astype(float)
```

```python
# CHECK DATA TYPES
print("Delivery[Avg customer Rating]:  ", delivery['Avg_Customer_Rating'].dtypes)
print("Delivery[Delivery Efficiency Score]:  ", delivery['Delivery_Efficiency_Score'].dtypes)
print("Restaurants[Avg Rating]:   ", restaurants['Avg_Rating'].dtypes)
```

```python
# VAlUE COUNT
print(delivery["Avg_Customer_Rating"].value_counts())
```

```python
delivery['Delivery_Efficiency_Score'].value_counts()
```

```python
restaurants['Avg_Rating'].value_counts()
```

```python
 # imputation
# Delivery Partners
delivery["Avg_Customer_Rating"].fillna(
    delivery["Avg_Customer_Rating"].median(), inplace=True
)

delivery["Delivery_Efficiency_Score"].fillna(
    delivery["Delivery_Efficiency_Score"].median(), inplace=True
)

# Restaurants
restaurants["Avg_Rating"].fillna(
    restaurants["Avg_Rating"].median(), inplace=True
)
```

```python
print(delivery["Avg_Customer_Rating"].median())
print(delivery["Delivery_Efficiency_Score"].median())
print(restaurants["Avg_Rating"].median())
```

```python
print(delivery["Avg_Customer_Rating"].median())
print(delivery["Delivery_Efficiency_Score"].mean())
print(restaurants["Avg_Rating"].median())
```

```python
print("Customers:  \n", customers.isnull().sum())
print("Delivery:  \n", delivery.isnull().sum())
print("Orders:    \n", orders.isnull().sum())
print("Restaurants:   \n", restaurants.isnull().sum())
```

```python
# 5. CHECK DUPLICATES
print("\nDuplicates:")
print("Customers:", customers.duplicated().sum())
print("Delivery:", delivery.duplicated().sum())
print("Orders:", orders.duplicated().sum())
print("Restaurants:", restaurants.duplicated().sum())
```

```python
# STANDARDIZE TEXT DATA
customers["City"] = customers["City"].str.strip().str.title()
customers["Customer_Type"] = customers["Customer_Type"].str.strip().str.title()
orders["Region"] = orders["Region"].str.strip().str.title()
orders["Payment_Mode"] = orders["Payment_Mode"].str.strip().str.title()
restaurants["City"] = restaurants["City"].str.strip().str.title()
```

```python
print("\nUnique Values Check:")
print("Customer Type:", customers["Customer_Type"].unique())
print("Payment Mode:", orders["Payment_Mode"].unique())
print("Region:", orders["Region"].unique())
```

```python
customers.to_excel("clean_customers.xlsx", index=False)
delivery.to_excel("clean_delivery.xlsx", index=False)
orders.to_excel("clean_orders.xlsx", index=False)
restaurants.to_excel("clean_restaurants.xlsx", index=False)
```
